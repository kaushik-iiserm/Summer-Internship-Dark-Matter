```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("plot-data.csv", header=0)
df1.columns
```

```python
df = pd.read_csv("gain.csv", header=0)
df.columns[2]
```

##### Gain Vs Drift(V/cm) [Avalanche Field = 400V/cm , Induction Field = 500 V/cm]

```python
df[(df['Avalanche'] == 400) & (df['Induction'] == 750)]
```

```python
gain = [40.8,40.64,40.64,42.87,42.9]
drfit = [100,250,500,750,1000]
effective_gain = [12.477,12.127,12.212,12.5,12.86]
error = np.array([1.32,0.86,0.91,0.95,0.88])
eff_error = effective_gain * error / gain
# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the theoretical gain
ax.errorbar(drfit, gain,yerr=error, fmt='-s',ecolor='blue',elinewidth=3,capsize=4, label='Theoretical Gain', markersize=3)

# Plot the effective gain
ax.errorbar(drfit, effective_gain, eff_error, fmt='-s',ecolor='blue',elinewidth=3,capsize=4, label='Effective Gain', markersize=3)

# Set the labels and title
ax.set_xlabel('Drift Field (V/cm)', fontsize=14)
ax.set_ylabel('Gain', fontsize=14)
ax.set_title('Gain vs Drift Field', fontsize=16, fontweight='bold')

# Set the y-axis limit
ax.set_ylim(0, 50)

# Add grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a legend
ax.legend(fontsize=12)

# Enhance the ticks
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', labelsize=10)

# Show the plot
plt.show()
```

```python
gain_a = [5.3,13,40,109,308]
avalanche = [300,350,400,450,500]
effective_gaina = [1.8,4.25,12.21,30.48,77.64]
error_a = np.array([0.16,0.56,1,2.14,3.76])
eff_error_a = effective_gaina * error_a / gain_a
# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(df1['x'],df1[' y'],'o',label='Experimental Data')
# Plot the theoretical gain
ax.errorbar(avalanche, gain_a, yerr=error_a, fmt='-s',ecolor='blue',elinewidth=3,capsize=4, label='Theoretical Gain', markersize=3)

# Plot the effective gain
ax.errorbar(avalanche, effective_gaina, yerr=eff_error_a,fmt='-s',ecolor='blue',elinewidth=3,capsize=4, label='Effective Gain', markersize=3)

# Set the labels and title
ax.set_xlabel('Avalanche Field (V/cm)', fontsize=14)
ax.set_ylabel('Gain', fontsize=14)
ax.set_title('Gain vs Avalanche Field', fontsize=16, fontweight='bold')

# Set the y-axis limit
ax.set_ylim(0, 350)

# Add grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a legend
ax.legend(fontsize=12)

# Enhance the ticks
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', labelsize=10)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
means =  df['Gain']
sigma_to_mean_ratio = df['sigma / mean ratio']
error_gain = df['error in gain']
error_sigma = df['error in spread']
# Plot
plt.figure(figsize=(14, 7))

# Scatter plot with error bars and transparency
plt.errorbar(means, sigma_to_mean_ratio, xerr=error_gain, yerr=error_sigma, fmt='^', ecolor='green',
             elinewidth=1, capsize=3, markersize=8, markerfacecolor='none', alpha=0.7, label='CO2/Ar')

# Adding a color bar to indicate sigma-to-mean ratio
sc = plt.scatter(means, sigma_to_mean_ratio, c=sigma_to_mean_ratio, cmap='viridis', alpha=0.7)
plt.colorbar(sc, label='Sigma-to-Mean Ratio')

# Adding gridlines
plt.grid(True, linestyle='--', alpha=0.6)

# Labels and title
plt.xlabel('GEM Gain')
plt.ylabel('$\sigma$/Mean')
plt.ylim(0.5, 1.2)
plt.title('Compilation of GEM response $\sigma$ / mean values')

# Legend
plt.legend()

# Display plot
plt.show()
```
