```python
import ROOT as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
gem_file = r.TFile.Open("GEM.root")
gem_file.ls()
tree = gem_file.Get("Garfield;1")
# tree.Print()
```

```python
initial_ion_times = []
for entry in tree:
    ion_times = getattr(entry, "initialIonTimes")
    for value in ion_times:
        initial_ion_times.append(value)
hist = plt.hist(initial_ion_times,bins=100, range=(0,4))
```

```python
avalanche_e = []
for entry in tree:
    avalanche = getattr(entry,"numberOfAvalancheElectrons")
    for value in avalanche:
        avalanche_e.append(value)
hist = plt.hist(avalanche_e,bins=100,range=(0,2000))
plt.title("number of Avalanche electrons")
np.mean(np.array(avalanche_e)), np.std(np.array(avalanche_e))
```

```python
detached_e_position_X = []
for entry in tree:
    position = getattr(entry,"initialElectronXpositions")
    for value in position:
        detached_e_position_X.append(value)
hist = plt.hist(detached_e_position_X,bins=100,range=(-0.006,0.006))
```
